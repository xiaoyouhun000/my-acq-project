```python
import pandas as pd
import datetime
import dateutil.parser as dparser
import pickle
import os
import re
```

```python
completepath = os.path.join("D:\Research\S&P Capital IQ","M&A data.csv")

acqdata = pd.DataFrame.from_csv(completepath,header = 0,index_col = None,parse_dates =['All Transactions Announced Date'])
acqdata = acqdata.drop_duplicates()
acqdata = acqdata[acqdata["CIK [Buyers/Investors]"]!= '-'].reset_index(drop = True)
```

```python
# find announcement filing

acqdata["8-K Filing"] = pd.Series(dtype = object)
acqdata["Other Form"] = pd.Series(dtype = object)
acqdata["Other Filing"] = pd.Series(dtype = object)

for i in range(len(acqdata)): #
    acqdate = acqdata.loc[i]["All Transactions Announced Date"]
    acqyear = acqdate.year
    acqquarter = acqdate.quarter
    buyercik = acqdata.loc[i]["CIK [Buyers/Investors]"]
    buyercik = int(re.search(r'\d+', buyercik).group())
    indexfile8k = "%s_%s.csv" %(acqyear,acqquarter)
    indexcsv8k = pd.DataFrame.from_csv(indexfile8k,header = 0,index_col = None,parse_dates = ['Date'])
    indexed_csv8k = indexcsv8k.set_index(['Date', 'CIK'])
    indexed_csv8k.sort_index(inplace=True)
    idx = pd.IndexSlice
    filings8k = []
    try:
        filings8k = indexed_csv8k.loc[idx[tuple(pd.date_range(acqdate-datetime.timedelta(days=1), periods=5).tolist()),buyercik],:]
    except:
        pass
    
    filing8k_url = []
    filingother_type = []
    filingother_url = []
    
    if len(filings8k)>0:
        for f in range(len(filings8k)):
            filing8k_url.append(filings8k.iloc[f]["url"])
        acqdata.set_value(i, '8-K Filing', filing8k_url)
    
    indexfileother = "%s_%s_all.csv" %(acqyear,acqquarter)
    indexcsvother = pd.DataFrame.from_csv(indexfileother,header = 0,index_col = None, parse_dates = ['Date'])
      
    indexed_csvother = indexcsvother.set_index(["Date","CIK"])
    indexed_csvother.sort_index(inplace = True)
    idx = pd.IndexSlice
    filingsother = []
    try:
        filingsother = indexed_csvother.loc[idx[tuple(pd.date_range(acqdate-datetime.timedelta(days=1), periods=5).tolist()),buyercik],:]
    except:
        pass
    if len(filingsother)>0:
        for f in range(len(filingsother)):
            filingother_type.append(filingsother.iloc[f]["Form Type"])
            filingother_url.append(filingsother.iloc[f]["url"])
        acqdata.set_value(i, 'Other Form', filingother_type)
        acqdata.set_value(i, 'Other Filing', filingother_url)


    
```
